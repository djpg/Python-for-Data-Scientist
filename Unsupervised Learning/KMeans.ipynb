# Evaluating Clustering 

A la hora de seleccionar cuál es el número de cluster, la mejor opción es la de calcular la inertia para distintos tipos de clúster y seleccionar un valor a partir del cual su valor comience a drecrecer lentamente.

```python
samples = np.loadtxt('samples.txt', delimiter = '\t')
ks = range(1,6)
inertias = []
for k in ks:
    #Nos creamos el modelo
    model = KMeans(n_clusters = k)
    #Fijamos el modelo
    model.fit(samples)
    #Calculamos la inertia para cada modelo
    inertias.append(model.inertia_)

_ = plt.plot(ks, inertias)
_ = plt.xlabel('number of clusters')
_ = plt.ylabel('inertia')
plt.show()
```

Podemos ver como el valor de 3 es una buena opción ya que a partir de este valor la inertia empieza a decrecer de forma lenta.

# Transforming features for better clusterings (wine)

Cuando tenemos variables en un dataset cuya varianza es muy elevada, pueden resultar un gran problema a la hora de asignarle el cluster adecuado. Es por esto, que es siempre recomendable cuando hacemos uso de clustering el de normalizar nuestras variables, es decir, ponerlas en el rango 0,1. La forma de normalizar nuestras variables depende de lo que busquemos por ejemplo la función StandardScaler lo que hace es eliminar la media y scalar la varianza a la unidad.

```python
#Importamos los datos
df = pd.read_csv('wine.csv')
y = df['class_name'].values
x = df.drop('class_name', axis = 1).values
#Creamos un modelo con tres clusters
model = KMeans(n_clusters = 3)
#Fijamos el modelo
labels = model.fit_predict(x)
#Nos creamos el dataframe
df_ct = pd.DataFrame({'labels': labels, 'class_name': y})
#Creamos la cross tab
ct = pd.crosstab(df_ct['labels'], df_ct['class_name'])
print(ct)
```

Podemos ver como obtenemos unos resultados bastante malos, es decir, ningún cluster unifica ninguna clase de vino. Esto es debido a que existen variables que tienen una desviación muy elevada.

```python
df.describe()
```

Podemos ver como la desviación estándar de la variable proline es 314, se trata de un valor excesivamente alto.

```python
#Importamos StandardScaler
from sklearn.preprocessing import StandardScaler
#Nos creamos el StandardScaler
scaler = StandardScaler()
#Fijamos sobre las muestras
scaler.fit(x)
#Aplicamos la transformación
x_scaled = scaler.transform(x)
```

Ahora deberiamos aplicar el algoritmo KMeans sobre estas muestras escaladas. Puesto que tendriamos que volver a realizar un fit y un transform la mejor forma de realizar esto para compactarlo todo es haciendo uso de los pipelines.

```python
from sklearn.pipeline import make_pipeline
#Nos creamos el scaler y el modelo 
scaler = StandardScaler()
model = KMeans(n_clusters = 3)
#Aplicamos el pipeline
pipeline = make_pipeline(scaler, model)
pipeline.fit(x)
labels = pipeline.predict(x)
#Nos creamos la cross tab
df_ct = pd.DataFrame({'labels':labels, 'class_names': y})
ct = pd.crosstab(df_ct['labels'], df_ct['class_names'])
print(ct)
```

Podemos ver claramente la mejora en nuestros resultados. Ahora el cluster 1 define de forma precisa el tipo de vino Grignolino. 

# Transforming features for better clusterings (fish)

Volveremos a repetir los pasos anteriores pero en este caso con un dataset diferente.

```python
#Importamos datos
df = pd.read_csv('fish.csv')
df.head()
```

```python
x = df.drop('Bream', axis = 1).values
y = df['Bream'].values
```

```python
#A continuación vamos a calcular el número de clusters óptimos
ks = range(1,10)
inertia = []
for k in ks:
    model = KMeans(n_clusters = k)
    model.fit(x)
    inertia.append(model.inertia_)
_ = plt.plot(ks, inertia)
_ = plt.xlabel('number of clusters')
_ = plt.ylabel('inertia')
plt.show()
```

Podemos ver que en este caso el número de cluster adecuado parece ser 4.

```python
#Nos creamos el scaler y el modelo
scaler = StandardScaler()
model = KMeans(n_clusters = 4)
#Seguidamente nos creamos el pipeline
pipeline = make_pipeline(scaler, model)
#Fijamos el pipeline
pipeline.fit(x)
#Predecimos los niveles
labels = pipeline.predict(x)
#Nos creamos el cross tab
df_ct = pd.DataFrame({'labels':labels, 'species':y})
ct = pd.crosstab(df_ct['labels'], df_ct['species'])
print(ct)
```

Podemos ver como estamos ante un resultado bastante bueno, la mayoría de los clusters identifican de forma clara a cada una de las especies de peces.

# Transforming features for better clusterings (stock)

A continuación vamos a proceder a hacer uso de KMeans para el conjunto de datos stock. El objetivo es agrupar distintas empresas en función de sus movimientos diarios, es decir, el precio de sus acciones (diferencia en dolares entre la hora a de apertura y la hora de cierre). Puesto que algunas acciones son más caras que otras hemos decidido de hacer uso de la función Normalizer, esta función a diferencia de StandardScaler, reescala cada muestra, del precio de las acciones de cada compañía independientemente de la otra.

```python
from sklearn.preprocessing import Normalizer
#Importamos los datos
df = pd.read_csv('stock.csv')
df.head()
```

```python
x = df.drop('Unnamed: 0', axis = 1).values
y = df['Unnamed: 0'].values
```

```python
#Calculamos el número de cluster necesarios
ks = range(5,20)
inertia = []
for k in ks:
    model = KMeans(n_clusters = k)
    model.fit(x)
    inertia.append(model.inertia_)
_ = plt.plot(ks, inertia)
_ = plt.xlabel('number of clusters')
_ = plt.ylabel('inertia')
plt.show()
```

En este caso podemos ver que el número de cluster más adecuado parece ser 13

```python
#Nos creamos el normalizer y el clustering
normalizer = Normalizer()
model = KMeans(n_clusters = 13)
#Nos creamos el pipeline
pipeline = make_pipeline(normalizer, model)
#Fijamos el pipeline
pipeline.fit(x)
#Predecimos
labels = pipeline.predict(x)
#Nos creamos dataframe
df = pd.DataFrame({'labels':labels, 'companies':y})
print(df.sort_values('labels'))
```

En este caso las compañías que tienen asignado el mismo clúster son compañías en las que sus acciones tienen tendencia a variar de la misma forma.
